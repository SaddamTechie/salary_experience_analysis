```python
# Load and Explore Data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = 'Employee_Salary_Dataset.csv'

data = pd.read_csv(filename)

print(data.head())
```

```python
print(data.info())
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Basic Data analysis
```

```python
data.describe()
```

```python
mean_salary_by_experience = data.groupby('Salary')['Experience_Years'].mean().sort_values(ascending=True)
print(mean_salary_by_experience)
```

```python
# Data Visualization
```

```python
# Plotting the line chart
plt.figure(figsize=(10,6))
plt.plot(mean_salary_by_experience.index, mean_salary_by_experience.values, marker='o')
plt.title('Average Salary by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary ($)')
plt.grid(True)
plt.xticks(mean_salary_by_experience.index) # Set x-ticks to be the years of experience
plt.show()
```

```python
# Plotting the bar chart
plt.figure(figsize=(10,6))
mean_salary_by_experience.plot(kind='bar', color='skyblue', edgecolor='black')

# Adding titles and labels
plt.title('Average Salary by Years of Experience', fontsize=16)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Average Salary ($)', fontsize=12)
plt.xticks(rotation=0) # Rotate x-ticks for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# Plotting the histogram for the 'Salary' column
plt.figure(figsize=(10,6))
data['Salary'].hist(bins=30, color='skyblue', edgecolor='black')

# Adding titles and labels
plt.title('Histogram of Salary Distribution', fontsize=16)
plt.xlabel('Salary ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
data.plot.scatter(x='Experience_Years', y='Salary', s=100, c='blue', alpha=0.5)

# Adding titles and labels
plt.title('Scatter Plot of Salary vs. Experience Years', fontsize=16)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```
